```python
'''
Hatfield 1/28/2018
Grabbing old code from the dsb 17 image viewer to repurpose for this year
Note: Last year required using DICOM images and constructing a slider to examine the 3d nature of the images
       which will not be required this year

'''

#Load necessary packages
import os
import numpy as np
import pandas as pd
from ipywidgets import widgets, interact, interactive, IntSlider
from IPython.display import display,clear_output,Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.patches import Circle
%matplotlib inline
```

```python
'''
Hard coded paths
set training folder path
'''

stage1_train = os.getcwd() + '\stage1_train'
#os.getcwd() gets you to the working directory, then navigate to the stage1_train folder
# this assuems that the .ipynb is located in the same directory as the stage1_train folder
```

```python
'''
Setting up the initial list of all image folders, along with a selector widget to choose individual ones to view
'''

image_names = sorted(os.listdir(stage1_train)) #list of all folder names
image_names_indexlist = [str(img_index) + ': ' for img_index in range(0, len(image_names))] #index for viewing purposes

combined_image_list = [x+y for x, y in zip(image_names_indexlist, image_names)] #combining the 2 lists above
image_names_list = widgets.Select(description = 'Images', options = combined_image_list) #creating the selector
#image_names_list = widgets.Select(description = 'Images', layout=Layout(width='100%', height='100%') ,options = combined_image_list) #creating the selector
```

```python
# global variables
selected_image = ''       #the chosen image from selector
image_path = ''           #full path for to image .png
masks = []                #list of masks for that image
masks_names_list = None   #widget for mask selection
img = None                #image plot for the full iamge

# function to load an images data
def load_image_data(args = ''):
    global selected_image
    global image_path
    global masks
    global masks_names_list
    
    selected_image = image_names[image_names_list.options.index(image_names_list.value)]
    image_path = stage1_train + '/' + selected_image
    image = image_path + '/images/' + selected_image + '.png'
    
    masks = os.listdir(image_path + '/masks')
    masks_indexlist = [str(img_index) + ': ' for img_index in range(0, len(masks))]
    combined_mask_list = [x+y for x, y in zip(masks_indexlist, masks)]
    if (masks_names_list is None): #If statement to check if we need to create the mask widget, or just change the data for it
        masks_names_list = widgets.Select(description = 'Masks', options = combined_mask_list) #creating mask selector
        masks_names_list.observe(plot_mask, 'value')
        display(masks_names_list)
    else:
        masks_names_list.options = combined_mask_list #changing mask selector data
        
    clear_output(wait = True) #clearing existing image figures
    plot_image(image)

#function to plot the image, image beeing the full view of nuclei
def plot_image(image):
    global img
    img = mpimg.imread(image)
    imgplot = plt.imshow(img)
    plt.show()

#function to plot the selected mask .png
def plot_mask(args = ''):
    global img
    clear_output(wait = True) #clearing exsting images
    
    #showing the full nuclei image
    imgplot = plt.imshow(img)
    plt.show()
    
    #showing the just the selected mask image
    selected_mask = masks[masks_names_list.options.index(masks_names_list.value)]
    mask_path = stage1_train + '/' + selected_image + '/masks/' + selected_mask
    img_mask = mpimg.imread(mask_path)
    imgplot = plt.imshow(img_mask)
    plt.show()
```

```python
#Set up the image selector and display it
image_names_list.observe(load_image_data, 'value')
display(image_names_list)
```
